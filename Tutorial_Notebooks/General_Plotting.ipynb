# Welcome to Spec_Tools

This python package is designed to help to analyze and visualize spectroscopic data. It provides a set of methods to read, process, plot and annotate spectroscopic data files. The package is designed to be easy to use and flexible.

```python
import sys
sys.path.append('..')
import spec_tools
```

## Data Generation

Spec_Tools provides different methods to import data from csv files, text files, and other formats. In this section of the tutorial scripts we generate random test data ourself and then try out different plotting methods contained in the **spectrum.py** module of the package.

For testing purposes the module **baseline.py** contains the function **generate_random_gaussian_peaks** which can be used to test the methods implemented in this tool

```python
import matplotlib.pyplot as plt
import numpy as np

x,y = spec_tools.baseline.generate_random_gaussian_peaks(10,[0,100],noise_level=0.5)


plt.plot(x,y)
```

```python
# Next of we generate a Spectrum Object from the data
data = np.array([x,y]).T

spectrum = spec_tools.spectrum.Spectrum(name="Test-Data", data=data)
```

### General Plotting

We can now plot the generated data using different plotting methods contained in the **spectrum.py** module of the package. These plots can be configurated using the dictionary initialized when generating the data this dictionary contains the following keys:

+ "x_label" : "Wavenumber / cm⁻¹",
+ "y_label" : "Intensity",
+ "axis_format" : "0.0f",
+ "interactive" : True,
+ "width" : 800,
+ "height" : 400,
+ "color" : "black",
+ "line_width" : 1,
+ "categorical_scheme" : "category10",
+ "color_reverse" : False,
+ "ppi" : 600,

```python
configuration_dictionary = spectrum.plot_configuration
print(configuration_dictionary)

spectrum.plot_spectrum(title="Test-Data")
```
